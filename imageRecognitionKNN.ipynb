```python
from PIL import Image
```

```python
length, height = 32, 32
```

```python
import numpy as np
```

```python
def img_deal(file):
    fl = open(file)
    matrix = np.zeros((1, length * height))
    for i in range(length):
        ln = fl.readline()
        for j in range(height):
            matrix[0, i*32+j] = int(ln[j])
            
    return matrix
```

```python
import operator
```

```python
def classify(single_test_data, train_data, labels, k):
    trainSize = train_data.shape[0]
    diffMatrix = np.tile(single_test_data, (trainSize, 1)) - train_data
    sqrMatrix = diffMatrix ** 2.
    sumMatrix = sqrMatrix.sum(axis = 1)
    #sqrtMatrix = sumMatrix ** 0.5
    inx = sumMatrix.argsort()
    count = {}
    for i in range(k):
        label = labels[inx[i]]
        val = count.setdefault(label, [0])
        val[0] += 1
    return max(count.items(), key = operator.itemgetter(1))[0]
```

```python
import os
```

```python
def parse_raw_file(file, subPath):
    currPath = os.getcwd()
    os.chdir(r"/home/niubuzhi/Documents/ml/image recognition using KNN/")
    fl = open(file)
    counter = dict((str(i), 0) for i in range(10))
    tmp = []
    while True:
        ln = fl.readline()
        if ln == '':
            break
        if ln[0] not in (' ', '0', '1'):
            continue
        elif ln[0] == ' ':
            output = subPath + ln[1] + '_' + str(counter[ln[1]]) + '.txt'
            counter[ln[1]] += 1
            with open(output, 'w') as of:
                for line in tmp:
                    of.write(line)
            tmp = []
        else:
            tmp.append(ln)
    os.chdir(currPath)
```

```python
parse_raw_file('optdigits-orig.tra', 'trainDigits/')
```

```python
currDir = os.getcwd()
os.chdir(r"/home/niubuzhi/Documents/ml/image recognition using KNN/trainDigits")

labels = []
all_files = os.listdir()
m_train = len(all_files)
train_data = np.zeros((m_train, length * height))
for i, fileName in enumerate(all_files):
    sp = fileName.split('_')
    label = sp[0]
    labels.append(label)
    train_data[i, :] = img_deal(fileName)

os.chdir(currDir)
```

### Now parse the test data

```python
parse_raw_file('optdigits-orig.cv', 'testDigits/')
```

### Now apply KNN to the test data

```python
currDir = os.getcwd()
os.chdir(r"/home/niubuzhi/Documents/ml/image recognition using KNN/testDigits")

correct = 0
all_files = os.listdir()
m_test = len(all_files)
for i, fileName in enumerate(all_files):
    sp = fileName.split('_')
    label = sp[0]
    test_data = img_deal(fileName)
    prediction = classify(test_data, train_data, labels, 3)
    if prediction == label:
        correct += 1
        print("Prediction is {}, Real is {}".format(prediction, label))

os.chdir(currDir)
```

```python
print('\nCorrect rate is {:.2f}%'.format(correct / m_test * 100))
```

### Now test a hand-written image

```python
currDir = os.getcwd()
os.chdir(r"/home/niubuzhi/Documents/ml/image recognition using KNN/handWritten")

img = Image.open('5_0.jpg')
img = img.resize((length, height))
img = img.convert('L')

test_written = np.zeros((1, length * height))
for i in range(height):
    for j in range(length):
        pixel = img.getpixel((j, i))
        if pixel > 170:
            test_written[0, 32*i + j] = 1
        else:
            test_written[0, 32*i + j] = 0

print(classify(test_written, train_data, labels, 3))
#img.show()

os.chdir(currDir)
```
